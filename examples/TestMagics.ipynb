# Query

```python
col1 = 'price'
col2 = 'bathrooms'
threshold = 2
```

## Short query

```python
price_in_area = %sql SELECT price, area FROM source_table
%show price_in_area -n 5
```

```python
price_bathrooms = %sql SELECT {col1}, {col2} FROM source_table
%show price_bathrooms -n 5
```
